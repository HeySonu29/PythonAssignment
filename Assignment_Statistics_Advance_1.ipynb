**Q3: Difference between discrete and continuous distributions**

The primary difference is that discrete distributions deal with countable, distinct outcomes, like the number of coin flips or dice rolls, while continuous distributions handle variables that can take any value within a given range, such as height or time. This distinction is reflected in how probabilities are assigned: discrete distributions use a probability mass function (PMF) to assign a probability to each specific value, whereas continuous distributions use a probability density function (PDF), where the probability of a specific value is zero, and probabilities are found by integrating the PDF over an interval.  

**Discrete Distributions**
Variables:
The random variable can only take on a finite or countable number of distinct values.
Examples:
The number of heads in 3 coin flips (0, 1, 2, or 3), the number of students in a class, or the result of a dice roll (1, 2, 3, 4, 5, or 6).
Probability:
Uses a Probability Mass Function (PMF) to assign a specific probability to each possible outcome.
The sum of all probabilities for all possible outcomes must equal 1.

**Continuous Distributions**
Variables: The random variable can take on any value within a given range.
Examples: A person's height, weight, the time it takes for a car to reach a certain speed, or the temperature of a room.
Probability:
Uses a Probability Density Function (PDF), which describes the relative likelihood for the variable to take on a given value.
The probability of a continuous variable being exactly equal to any single value is zero.
Probabilities are calculated by finding the area under the PDF curve over a specific interval.

```python
"""Q9: Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
deviation using NumPy, and draw a histogram to visualize the distribution."""

import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers from normal distribution
data = np.random.normal(50, 5, 1000)

# Compute mean and std deviation
mean = np.mean(data)
std_dev = np.std(data)

print("Mean:", mean)
print("Standard Deviation:", std_dev)

# Draw histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.title("Normal Distribution (mean=50, std=5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
"""Question 10: You are working as a data analyst for a retail company. The company has
collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
‚óè Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.
‚óè Write the Python code to compute the mean sales and its confidence interval."""

import numpy as np
import scipy.stats as st

# Given daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Compute mean and standard error
mean_sales = np.mean(daily_sales)
sem = st.sem(daily_sales)

# 95% confidence interval
confidence_interval = st.t.interval(0.95, len(daily_sales)-1, loc=mean_sales, scale=sem)

print("Mean Sales:", mean_sales)
print("95% Confidence Interval:", confidence_interval)
```
