**Question 1: What is hypothesis testing in statistics?**

Hypothesis testing is a formal **statistical procedure** used to make decisions about a population based on sample data. It helps determine whether the observed results are due to random chance or represent a real effect.


**Steps in Hypothesis Testing:**
1. **Formulate hypotheses:**
- Null Hypothesis (H₀): No effect or no difference.
- Alternative Hypothesis (H₁): There is an effect or difference.
2. **Select significance level (α):** Usually 0.05 or 0.01.
3. **Choose test statistic:** Z, t, χ², etc.
4. **Compute test statistic:** Based on sample data.
5. **Decision Rule:** Compare p-value with α or test statistic with critical value.
6. **Conclusion:** Reject or fail to reject H₀.


**Example:** Testing whether the average weight of apples in an orchard is 100g.
- H₀: μ = 100
- H₁: μ ≠ 100
If the calculated p-value < 0.05, we reject H₀ and conclude the average weight is significantly different from 100g.

### Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

**Answer:**
- **Null Hypothesis (H₀):** Represents the default assumption that there is no significant difference or effect. It is what we attempt to test against.
- **Alternative Hypothesis (H₁):** Represents the claim we want to prove; it indicates a significant difference or effect.


| Aspect | Null Hypothesis (H₀) | Alternative Hypothesis (H₁) |
|-------|---------------------|---------------------------|
| Meaning | No effect, no difference | Some effect, some difference |
| Goal | Try to disprove | Try to prove |
| Example | Mean weight = 70 kg | Mean weight ≠ 70 kg |


**Key Idea:** Hypothesis testing is a way to see if data provides enough evidence to reject H₀ in favor of H₁.

### Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

**Answer:**
The **significance level (α)** is the probability threshold used to decide whether to reject the null hypothesis. It represents the risk of making a **Type I error**.


**Formula:**
\[ \alpha = P(\text{Reject H₀ | H₀ is true}) \]


**Typical Values:**
- α = 0.05 (5%) → common in most tests.
- α = 0.01 (1%) → stricter, used when errors are costly.


**Role:**
- If **p-value ≤ α** → reject H₀ (result is statistically significant).
- If **p-value > α** → fail to reject H₀ (no sufficient evidence).


**Example:** In medical trials, α = 0.01 may be chosen to minimize false positives when testing a new drug.

### Question 4: What are Type I and Type II errors? Give examples of each.

**Answer:**
| Error Type | Definition | Example |
|-----------|------------|---------|
| **Type I Error (α)** | Rejecting a true null hypothesis (false positive) | Declaring a healthy person sick in a medical test |
| **Type II Error (β)** | Failing to reject a false null hypothesis (false negative) | Declaring a sick person healthy |


**Relationship:**
- Type I error is controlled by significance level α.
- Power of a test = 1 − β (probability of correctly rejecting false H₀).


Visualizing this using a courtroom analogy:
- Type I error = convicting an innocent person.
- Type II error = letting a guilty person go free.

### Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

**Answer:**
| Aspect | Z-Test | T-Test |
|-------|--------|--------|
| Population Variance | Known | Unknown |
| Sample Size | Large (n > 30) | Small (n ≤ 30) |
| Distribution Used | Normal Distribution | Student’s t-Distribution |
| Application | Population mean, proportions | Population mean |


**Formula (Z):**
\[ Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}} \]


**Formula (t):**
\[ t = \frac{\bar{X} - \mu}{s / \sqrt{n}} \]
where s = sample standard deviation.


**Example:**
- Z-test: Quality control in a factory where population variance is known.
- T-test: Comparing mean exam scores of a small class with a standard.

```python
### Question 7: Hypothesis testing using Z-statistics for a sample dataset.

# Answer:
import numpy as np
from scipy import stats


# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]


pop_mean = 50
pop_std = 0.5
sample_mean = np.mean(sample_data)


z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(len(sample_data)))
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))


print("Sample Mean:", sample_mean)
print("Z-statistic:", z_stat)
print("p-value:", p_value)

# Interpretation: If p-value < 0.05, reject H₀ (sample mean is significantly different). If p-value > 0.05, fail to reject H₀ (no significant difference).
```

```python
### Question 8: Simulate data from a normal distribution and calculate the 95% confidence interval.

#Answer:

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats


data = np.random.normal(loc=50, scale=5, size=1000)
mean = np.mean(data)
sem = stats.sem(data)
conf_interval = stats.norm.interval(0.95, loc=mean, scale=sem)


print("Mean:", mean)
print("95% CI:", conf_interval)


plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.title('Normal Distribution Simulation')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
### Question 9: Calculate Z-scores and visualize.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats


def z_score_visualization(data):
    z_scores = stats.zscore(data)
    plt.hist(z_scores, bins=30, edgecolor='black', alpha=0.7)
    plt.title('Z-score Distribution')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.show()
    return z_scores


# Example usage
data = np.random.normal(50, 5, 1000)
z_scores = z_score_visualization(data)

# Explanation: Z-scores measure how far a data point is from mean in standard deviations. Z=0 means the data point is at the mean, Z>0 means the data point is above the mean, and Z<0 means the data point is below the mean.
```
